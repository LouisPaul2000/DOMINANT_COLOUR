```
import cv2
import matplotlib.pyplot as plt
import numpy as np
!pip install opencv-python
from sklearn.cluster import KMeans
```

```
img = cv2.imread("SELF.jpg")
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB) 

print(img.shape)
print("SHIVANSH")
```

```
plt.imshow(img)
```

```
all_pixel = img.reshape((-1,3))
print(all_pixel.shape)
```

```
dominant_color = 4

km = KMeans(n_clusters=dominant_color)
```

```
km.fit(all_pixel)
```

```

centers = km.cluster_centers_
print(centers)
```

```
centers = np.array(centers , dtype = "uint16")
```

```
centers
```

```
np.unique(km.labels_ , return_counts= True)
```

```

new_image = np.zeros((img.shape[0]*img.shape[1] , 3) , dtype="uint16")
```

```
centers
```

```
for i in range(new_image.shape[0]):
    new_image[i] = centers[km.labels_[i]]
```

```
new_image = new_image.reshape((img.shape[0] , img.shape[1] , 3))
```

```
plt.imshow(new_image)
```
